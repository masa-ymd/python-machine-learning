## クラスタ分析

クラスタ分析は教師なし学習に属し、正しい答えが事前にわかっていないデータから隠れた構造を見つけ出す手法である<br>
クラスタ分析のカテゴリは以下の３つ
* プロトタイプベースクラスタリング<br>
  実装がシンプルで、他のアルゴリズムと比べて計算効率が非常に良いことがメリットとして挙げられる<br>
  また、クラスタ内のサンプルの分布が円、球状になっていることが前提になっている<br>
  クラスタの中心がセントロイド（特徴量が連続値の場合）かメドイド（特徴量がカテゴリ値の場合）のどちらかになる 
* 階層的クラスタリング<br>
  メリットとして樹形図をプロットし、結果の解釈に役立てることができる。また事前にクラスタの個数を指定する必要がない<br>
  大別して以下の２つの手法がある
  * 分割型…まず全てのサンプルを含むクラスタを定義し、1サンプルのみを含むクラスタになるまで分割する
  * 凝集型…上と逆で、1つのサンプルのみのクラスタをマージしていき、1つのクラスタにまとまるまでマージを繰り返す<br>
    凝集型のクラスタ間の距離を決めるアルゴリズムには以下の4つがある
    * 単連結法…各クラスタ内のサンプルの内、最も近いものをクラスタ間の距離とする
    * 完全連結法…上と逆で、最も遠いものをクラスタ間の距離とする
    * 平均連結法…クラスタ内の各サンプルの最小距離の平均をクラスタ間の距離とする
    * ウォード連結法…マージした場合のクラスタ内のSSE増加量をクラスタ間の距離とする
* 密度ベースクラスタリング<br>
  局所的な密度に基づいて点をグループ化するアルゴリズム<br>
  密度ベースクラスタリングの1つであるDBSCANは外れ値を処理できるほか、円・球状以外の形状の識別を行うことができる

## プロトタイプベースクラスタリング

### k-means法

上記のプロトタイプベースクラスタリングの１つで、産業界でも学術界でも広く使用されている<br>
具体的な方法は以下
1. サンプル点からk個の初期値をランダムに選択し、クラスタの中心（セントロイド）の初期値とする
2. 各サンプルをもっとも近いセントロイド$\mu^{(i)}$と同じクラスタに所属させる
3. クラスタの中心になるサンプルを新たなセントロイドとする
4. クラスタが変化しなくなるか、ユーザの指定した閾値、あるいはイテレーションの最大回数に達するまで1〜3を繰り返す

手順2の「最も近い」を図る指標としてよく使われるのは、以下に示すユークリッド距離の2乗
$$
d(x, y)^2 = \sum_{j=1}^m (x_j - y_j)^2 = \| {\bf x} - {\bf y} \|_2^2
$$
※　$\|$はノルム（距離）を表す<br>
※　下付きの数字2は$L^2$ノルムであることを表す。$L^p$ノルムは$\sqrt[p]{|x_1|^p + |x_2|^p + \cdots + |x_m|^p}$<br>
ユークリッド距離を用いる場合、k-means法で最小化するのはクラスタ内誤差平方和（SSE)…クラスタの中心からの離れ具合
$$
SSE = \sum_{i=1}^{n}\sum_{j=1}^k w^{(i,j)} \| {\bf x^{(i)}} - {\bf \mu^{(i)}} \|_2^2
$$
※　iはサンプルのインデクス<br>
※　jはクラスタのインデックス<br>
※　$\mu^{(i)}$はクラスタの中心<br>
※　サンプルがクラスタ内に存在するときは$w^{(i, j)} = 1$、そうでない時は0<br>

```python
# データセットを準備
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150,   # サンプル数
                  n_features=2,    # 特徴量の数
                  centers=3,       # クラスタの個数
                  cluster_std=0.5, # クラスタ内の標準偏差
                  shuffle=True,    # サンプルをシャッフル
                  random_state=0   # 乱数生成器の状態を指定
                 )
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c='white', marker='o', s=50)
plt.grid()
plt.show()
```

```python
# scikit-learnのkmeansを利用する例
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,   # クラスタの数
            init='random',  # セントロイドの初期値の選択方法（ランダムに選択）
            n_init=10,      # 異なるセントロイドを用いたアルゴリズムの実行回数（もっともSSEが小さいものを最終モデルとして選択）
            max_iter=300,   # 各セントロイド毎の試行回数
            tol=1e-04,      # 収束判定を行うための誤差（SSEが0.0001以下）
            random_state=0  # セントロイドの初期化に用いる乱数生成器の状態
           )
y_km = km.fit_predict(X)
```

### k-means++法

k-means法では乱数を使ってセントロイドの初期値を設定するが、初期値が不適切な場合に、クラスタリングがうまく行かなかったり、<br>
収束に時間がかかったりすることがある<br>
k-means++法では、以下の手順で初期のセントロイドを互いに離れた位置に配置することでこの問題に対処する
1. セントロイド$\mu^(i)$をランダムに１つ選択する
2. セントロイドではないサンプルごとに、セントロイドに選択されたサンプルとの距離$d({\bf x^{(i)}}, {\bf M})$が最小になるものを求める<br>
※　Mはセントロイドに選択されたサンプル群
3. 次のセントロイド$\mu^{(p)}$を選択する。選択は、各サンプルの距離の重みを等しくした以下の確率分布に従って行う<br>
$$
\frac{d({\bf \mu^{(p)}}, {\bf M})^2}{\sum_i d({\bf x^{(i)}, {\bf M}})^2}
$$
※　各セントロイドから離れていると選ばれやすくなる
4. k個のセントロイドが選ばれるまで2〜3を繰り返す
5. 従来のk-means法を使って引き続き処理を行う

```python
# k-means++法を実行
km = KMeans(n_clusters=3,   # クラスタの数
            init='k-means++',  # セントロイドの初期値の選択方法（k-means++法を指定）
            n_init=10,      # 異なるセントロイドを用いたアルゴリズムの実行回数（もっともSSEが小さいものを最終モデルとして選択）
            max_iter=300,   # 各セントロイド毎の試行回数
            tol=1e-04,      # 収束判定を行うための誤差（SSEが0.0001以下）
            random_state=0  # セントロイドの初期化に用いる乱数生成器の状態
           )
y_km = km.fit_predict(X)
```

```python
# 結果をプロット
plt.scatter(X[y_km==0, 0], X[y_km==0, 1], s=50, c='lightgreen', marker='s', label='cluster 1')
plt.scatter(X[y_km==1, 0], X[y_km==1, 1], s=50, c='orange', marker='o', label='cluster 2')
plt.scatter(X[y_km==2, 0], X[y_km==2, 1], s=50, c='lightblue', marker='v', label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=250, marker='*', c='red', label='centroids')
plt.legend()
plt.grid()
plt.show()
```

k-meansのもうひとつの問題は空になるクラスタが存在する可能性があることだが、scikit-learnの実装ではクラスタが空の場合、
そのクラスタから最も離れているサンプルを探してセントロイドにすることで問題に対処している

#### ハードクラスタリングとソフトクラスタリング

* ハードクラスタリング…サンプルが1つのクラスタに割り当てられるアルゴリズム<br>
サンプルの所属関係を以下のようなベクトルで表現できる
$$
\left[ \begin{array}{r} \mu^{(1)} \to 0 \\  \mu^{(2)} \to 1 \\  \mu^{(3)} \to 0 \\\end{array} \right]
$$
* ソフトクラスタリング…サンプルを1つ以上のクラスタに割り当てるアルゴリズム<br>
サンプルの所属関係を以下のようなベクトルで表現できる
$$
\left[ \begin{array}{r} \mu^{(1)} \to 0.1 \\  \mu^{(2)} \to 0.85 \\  \mu^{(3)} \to 0.05 \\\end{array} \right]
$$
※ 上記の数値は0以上1以下で、合計が1になる

#### FCM法

ソフトクラスタリングの１種であるFCM法の手順は以下の通り
1. セントロイドの個数kを決め、各サンプル点において、k個のクラスタに属する確率をランダムに割り当てる
2. 各クラスタのセントロイドを計算する
3. 各サンプル点のクラスタに属する確率を更新する
4. 確率が変化しなくなるか、最小化する目的関数の値が閾値未満の値になるか、繰り返しの最大回数に達するまで2〜3を繰り返す

目的関数$J_m$は以下の式で表される
$$
j_m = \sum_{i=1}^n \sum_{j=1}^k w^{(i, j)^m} \| {\bf x^{(i)}} - {\bf \mu^{(i)}}\|_2^2, m \in [1, \infty)
$$
※　$w^{(i, j)}$はクラスタに属する確率<br>
※　mはファジー係数で、mが大きくなるほど$w^{(i, j)}$は小さくなり、より範囲がぼやけるクラスタとなる<br>
※　$w^{(i, j)}$は以下の式で表せる
$$
w^{(i, j)} = \left[ \sum_{p=1}^k \left( \frac{\| {\bf x^{(i)}} - {\bf \mu^{(j)}} \|_2}{\| {\bf x^{(i)}} - {\bf \mu^{(p)}} \|_2} \right)^{\frac{2}{m-1}} \right]^{-1}
$$
※　クラスタの中心$\mu^{(j)}$はそのクラスタの全てのサンプルの平均として計算され、以下の式で表せる
$$
\mu^{(j)} = \frac{\sum_{i=1}^n w^{(i,j)^m}{\bf x^{(i)}}}{\sum_{i=1}^n w^{(i,j)^m}}
$$

### クラスタリングの評価方法

#### エルボー法

クラスタ解析は教師なし学習のため明確な答えがわからないが、例えばクラスタ内SSE(歪み)を指標とし、パラメータの異なるk-meansの性能を比較する<br>
ことができる。エルボー法は、kの値が増えれば、各サンプルが適切なセントロイドに近づくため、歪みが減るという考えに基づき、<br>
kの変化に伴うSSEをプロットし、歪みが大きく改善されるkの値を特定する方法である

```python
# SSEを表示
print('Distorsion: %.2f' % km.inertia_)
```

```python
# エルボー法
distorsions = []
for i in range(1, 11): # kを1から10まで試す
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    km.fit(X)
    distorsions.append(km.inertia_)
    
plt.plot(range(1, 11), distorsions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

上記から、ちょうど肘の位置にあるk=3が効果的なkの値であることが分かる

#### シルエット分析

以下の手順で、クラスタ内のサンプルがどの程度密にグループ化されているか（凝集度）の目安となるグラフをプロットできる
1. サンプル$x^{(i)}$と同じクラスタ内の他のサンプルとの平均距離を、クラスタの凝集度$a^{(i)}$として計算する
2. サンプル$x^{(i)}$と最も近いクラスタの全てのサンプルの平均距離を、クラスタからの乖離度$b^{(i)}$として計算する
3. 以下の式でシルエット係数$s^{(i)}$を計算する
$$
s^{(i)} = \frac{b^{(i)} - a^{(i)}}{max\{ b^{(i)}, a^{(i)} \}}
$$

※　シルエット係数の範囲は-1から1まで<br>
※　乖離度 > 凝集度の場合は理想的なシルエット係数である1に近づく<br>
※　乖離度 << 凝集度の場合はシルエット係数は-1に近づき、この時サンプル$x^{(i)}$は所属するクラスタより最も近いクラスタに近いことを表す<br>

```python
# k=3の場合のシルエット係数をプロット
km = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)

import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

# シルエット係数を計算
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []

# シルエット図を描画
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort() # 昇順に並び替え
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', linestyle='--')
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.show()
```

横軸がシルエット係数で、理想となる1からは離れていることが分かる<br>
また、縦軸の各クラスタ毎の幅がサンプル数を表すが、均等に分かれていることが読み取れる<br>
より悪いクラスタ分割の場合、シルエット図がどうなるかの例を以下に示す。

```python
# k=2の例
km = KMeans(n_clusters=2,   # クラスタの数
            init='k-means++',  # セントロイドの初期値の選択方法（k-means++法を指定）
            n_init=10,      # 異なるセントロイドを用いたアルゴリズムの実行回数（もっともSSEが小さいものを最終モデルとして選択）
            max_iter=300,   # 各セントロイド毎の試行回数
            tol=1e-04,      # 収束判定を行うための誤差（SSEが0.0001以下）
            random_state=0  # セントロイドの初期化に用いる乱数生成器の状態
           )
y_km = km.fit_predict(X)
plt.scatter(X[y_km==0, 0], X[y_km==0, 1], s=50, c='lightgreen', marker='s', label='cluster 1')
plt.scatter(X[y_km==1, 0], X[y_km==1, 1], s=50, c='orange', marker='o', label='cluster 2')
plt.scatter(X[y_km==2, 0], X[y_km==2, 1], s=50, c='lightblue', marker='v', label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=250, marker='*', c='red', label='centroids')
plt.legend()
plt.grid()
plt.show()
```

```python
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

# シルエット係数を計算
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []

# シルエット図を描画
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort() # 昇順に並び替え
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', linestyle='--')
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.show()
```

上記から分かるように、シルエット係数とクラスタ内のサンプルに大きなばらつきがあり、最適なクラスタリングでないことが裏付けられている

## 階層的クラスタリング

例として完全連結法に基づく、凝集型階層的クラスタリングをあげる<br>
手順は以下の通り<br>
1. 全てのサンプルの距離行列を計算する
2. 各データを１つのクラスタとみなす
3. 最も離れたメンバーの距離をクラスタ間の距離とし、２つの最も距離が近いクラスタをマージする
4. 距離行列を更新する
5. クラスタが1つになるまで3,4を繰り返す

```python
# テストデータの用意
import pandas as pd
import numpy as np
np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1', 'ID_2', 'ID_3', 'ID_4']
X = np.random.random_sample([5, 3])*10 # 5行3列のサンプルデータを作成
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
# 手順1
from scipy.spatial.distance import pdist, squareform
# pdistでサンプル間の距離を計算、squareformで対象行列を作成
dist = pdist(df, metric='euclidean')
dist
```

```python
square = squareform(dist)
square
```

```python
row_dist = pd.DataFrame(square, columns=labels, index=labels)
row_dist
```

```python
# 手順3-5
from scipy.cluster.hierarchy import linkage
row_clusters = linkage(df.values, method='complete', metric='euclidean')
row_clusters
```

```python
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2', 'distance', 'no. of items in clust.'],
             index=['cluster %d' %(i+1) for i in range(row_clusters.shape[0])]
            )
```

row label 1,2が各サンプルを表す<br>
5, 6, 7はマージしてできたものを指す<br>
distanceはクラスタ間の距離、no. of itemsはクラスタ内のサンプル数を表す

```python
# 樹形図を表示
from scipy.cluster.hierarchy import dendrogram
row_dendr = dendrogram(row_clusters, labels=labels)
plt.ylabel('Euclidean distance')
plt.show()
```

```python
# 樹形図をヒートマップと合わせて表示する
fig = plt.figure(figsize=(8,8), facecolor='white') # 新しいfigureオブジェクト定義
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6]) # 樹形図用の軸の位置、サイズ指定
row_dendr = dendrogram(row_clusters, orientation='left') # 樹形図を90度半時計回りに回転させて描画
df_rowclust = df.ix[row_dendr['leaves'][::-1]] # ixでデータフレームのデータを行列指定で選択可能、leavesで樹形図の末端の値取得、:: で指定した数スキップ、これで末端の数値を樹形図と同じ順番で取得する
axm = fig.add_axes([0.23, 0.1, 0.6, 0.6]) # ヒートマップ用の軸、サイズ指定
cax = axm.matshow(df_rowclust, interpolation='nearest', cmap='hot_r') # ヒートマップを描画
axd.set_xticks([]) # 目盛りを消す
axd.set_yticks([]) # 目盛りを消す
for i in axd.spines.values(): # 樹形図の枠線を消す
    i.set_visible(False)
fig.colorbar(cax) # ヒートマップにカラーバーをつける
axm.set_xticklabels([''] + list(df_rowclust.columns)) # ヒートマップの横軸のラベルを特徴量にする
axm.set_yticklabels([''] + list(df_rowclust.index)) # ヒートマップの縦軸のラベルをサンプル名にする
plt.show()
```

樹形図のクラスタの距離をとヒートマップの色の近さがあっており、視覚的にわかりやすい？

```python
# scikit-leanの凝集型階層的クラスタリング
# クラスタの剪定を行うことができる
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=2,          # クラスタの個数（剪定）
                             affinity='euclidean',  # 類似度の指標を指定
                             linkage='complete'     # 連結方法
                            )
labels = ac.fit_predict(X)
print('Cluster labels: %s' % labels)
```

上記はscipyの樹形図を上から2段めで切った結果と同じになっている

## 密度ベースクラスタリング

#### DBSCAN (density based spatial clustering of  applications with noise)

DBSCANは以下の条件に基づき、各サンプルにラベルが割り当てられる
* 指定された半径$\epsilon$以内に、少なくとも指定された個数（MinPts）の隣接点があるような点は「コア点」とみなされる
* 半径$\epsilon$以内の隣接点の個数が、MinPts未満だが、コア点の半径$\epsilon$内に位置するものは「ボーダー点」とみなされる
* 上記の２つに当てはまらないものは「ノイズ点」とみなされる

ラベル割りあてのあとの処理は以下の2ステップにまとめられる
1. コア点ごとに、またはコア点の接続関係に基づいて、別々のクラスタを形成する（コア点が$\epsilon$よりも離れていない時、接続関係を持つ）
2. 各ボーダー点をそれと対になっているコア点のクラスタに割り当てる

```python
# テストデータを生成
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# k-meansと階層型クラスタリングで上記を分離できるか確認
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

# k-means
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)
ax1.scatter(X[y_km==0, 0], X[y_km==0, 1], c='lightblue', marker='o', s=40, label='cluster 1')
ax1.scatter(X[y_km==1, 0], X[y_km==1, 1], c='red', marker='s', s=40, label='cluster 2')
ax1.set_title('K-means clustering')

# 階層型クラスタリング
ac = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
y_ac = ac.fit_predict(X)
ax2.scatter(X[y_ac==0, 0], X[y_ac==0, 1], c='lightblue', marker='o', s=40, label='cluster 1')
ax2.scatter(X[y_ac==1, 0], X[y_ac==1, 1], c='red', marker='s', s=40, label='cluster 2')
ax2.set_title('Agglometative clustering')
plt.legend()
plt.show()
```

上記の結果から、どちらもうまくクラスタリングできていないことが分かる

```python
# DBSCANでクラスタリングを試す
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.2,           # 隣接点とみなす２点間の最大距離
            min_samples=5,     # ボーダー点の最小個数(MinPts)
            metric='euclidean' # 距離の計算法
            )
y_db = db.fit_predict(X)

plt.scatter(X[y_db==0, 0], X[y_db==0, 1], c='lightblue', marker='o', s=40, label='cluster 1')
plt.scatter(X[y_db==1, 0], X[y_db==1, 1], c='red', marker='s', s=40, label='cluster 2')
plt.legend()
plt.show()
```

DBSCANにより半月状のクラスタをうまく検出できることができることがわかる
