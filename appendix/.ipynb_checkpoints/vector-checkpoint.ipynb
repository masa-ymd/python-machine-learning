# 行列の固有値分解の基礎

### 行列によるベクトルの回転

以下の行列Aをベクトルにかけて、ベクトルを回転・拡大させることを考える
$$
A = \left(
  \begin{array}{rr}
    1 & 1 \\
    -2 & 4 \\
  \end{array}
\right)
$$

回転させるベクトルは以下の2つ
$$
x_1 = \left( \begin{array} {r} 1\\ 0\\ \end{array} \right), x_2 = \left( \begin{array} {r} 0\\ 1\\ \end{array} \right)
$$

```python
# 回転前のベクトルをプロットする
%matplotlib inline
import matplotlib.pyplot as plt
# LaTexのフォント指定
#plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams['text.latex.preamble'] = [r"/usepackage{amsmath}"]
# フォントサイズ指定
plt.rcParams['font.size'] = 18

# ベクトルを描画
import numpy as np
x1 = np.array([[1], [0]])
x2 = np.array([[0], [1]])
plt.quiver(0, 0, np.hstack((x1, x2))[0, :], np.hstack((x1, x2))[1, :],
         color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
plt.text(0.5, -0.5, '${x}_{1}$')
plt.text(-0.7, 0.5, '${x}_{2}$')

# 軸の範囲
plt.xlim(-3, 3)
plt.ylim(-3, 4)
# 軸の目盛りを設定
plt.xticks(np.arange(-3, 3.1, 1))
plt.yticks(np.arange(-3, 4.1, 1))
plt.grid()
ax = plt.gca()
# 縦横軸の目盛りの大きさを合わせる
ax.set(adjustable='box-forced', aspect='equal')
plt.show()
```

それぞれのベクトルに行列をかけると以下のようになる
$$
y_1 = Ax_1 = 
\left(
  \begin{array}{rr}
    1 & 1 \\
    -2 & 4 \\
  \end{array}
\right)
\left(
  \begin{array}{r}
    1 \\
    0 \\
  \end{array}
\right)
=
\left(
  \begin{array}{rr}
    1 \times 1 + 1 \times 0 \\
    -2 \times 1 + 4 \times 0 \\
  \end{array}
\right)
=
\left(
  \begin{array}{r}
    1 \\
    -2 \\
  \end{array}
\right)
$$
$$
y_2 = Ax_2 = 
\left(
  \begin{array}{rr}
    1 & 1 \\
    -2 & 4 \\
  \end{array}
\right)
\left(
  \begin{array}{r}
    0 \\
    1 \\
  \end{array}
\right)
=
\left(
  \begin{array}{rr}
    1 \times 0 + 1 \times 1 \\
    -2 \times 0 + 4 \times 1 \\
  \end{array}
\right)
=
\left(
  \begin{array}{r}
    1 \\
    2 \\
  \end{array}
\right)
$$

```python
# Numpyで演算を行う
A = np.array([[1, 1], [-2, 4]])
y1, y2 = A.dot(x1), A.dot(x2)
print(y1)
print(y2)
```

```python
# x1とx2を列方向に結合してAをかけることで、各列が変換されたベクトルになる
y = A.dot(np.hstack((x1, x2)))
print(y)
```

```python
# 行列Aを掛けたことによるベクトルを図示する
fig = plt.figure()
plt.quiver(0, 0, np.hstack((y1, y2))[0, :], np.hstack((y1, y2))[1, :],
           color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
plt.text(0.8, -1.5, '${x}_{1}$')
plt.text(-0.3, 3.0, '${x}_{2}$')
plt.xlim(-3, 3)
plt.ylim(-3, 4.5)
plt.xticks(np.arange(-3, 3.1, 1))
plt.yticks(np.arange(-3, 4.1, 1))
plt.grid()
ax = fig.gca()
ax.set(adjustable='box-forced', aspect='equal')
plt.show()
```

### 固有ベクトル（行列をかけても向きが変化しないベクトル）

```python
# -1から1まで0.005刻みでグリッドポイント生成
x, y = np.meshgrid(np.arange(-1, 1.01, 0.005), np.arange(-1, 1.01, 0.005))
# x, yを一次元配列に直して行方向vに結合しAをかける
xx, yy = A.dot(np.vstack((x.ravel(), y.ravel())))
# 元のベクトルの長さ
norm_orig = np.sqrt(x.ravel()**2 + y.ravel()**2)
# 変換後のベクトルの長さ
norm_mapped = np.sqrt(xx**2 + yy**2)
# ベクトルの内積
inner_prod = xx * x.ravel() + yy * y.ravel()
# 逆三角関数の余弦(y = arccos x <-> x = cos y)
# cosθ = ( AとBの内積 ) / (Aの長さ * Bの長さ)
theta = np.arccos(inner_prod / (norm_orig * norm_mapped))
# ベクトル間の角度
interval = np.arange(0, max(theta) + 0.05, 0.01)
plt.contourf(x, y, theta.reshape(x.shape), interval)
# カラーバーを表示
plt.colorbar()
plt.grid()
plt.show()
```

上記の濃い青の部分（2本の線）が2つのベクトルのなす角が0、すなわち行列をかけても角度が変化しないベクトル
この性質は以下のように表せる
$$
\bf {A = \lambda p}
$$
この時、ベクトルpを行列Aの固有ベクトル、$\lambda$を行列Aの固有値と呼ぶ

```python
eigval_A, eigvec_A = np.linalg.eig(A)
print(eigval_A)
print(eigvec_A)
```

上記の結果は、$\lambda_1=2, \lambda_2=3$、固有ベクトルは
$$
p_1 = \left(
    \begin{array}{r}
      -0.70710678 \\
      -0.70710678 \\
    \end{array}
  \right),
p_2 = \left(
    \begin{array}{r}
      -0.4472136 \\
      -0.89442719 \\
    \end{array}
  \right)
$$
と求められている
$$
Ap_1 = \lambda_1p_1, Ap_2 = \lambda_2p_2
$$
※固有値、固有ベクトルは固有方程式から解析的に求めることができる

```python
# 固有ベクトル、固有値を可視化する
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(8, 8))
ax[0].quiver(0, 0, eigvec_A[0, :], eigvec_A[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[0].text(-0.7, -0.4, '${p}_{1}$')
ax[0].text(-0.2, -0.8, '${p}_{2}$')
ax[0].set_xlim(-2, 1.5)
ax[0].set_ylim(-3, 0.5)
ax[0].set_xticks(np.arange(-2, 1.1, 1))
ax[0].set_yticks(np.arange(-3, 0.1, 1))
ax[0].set_title('original')
ax[0].set(adjustable='box-forced', aspect='equal')
ax[0].grid()
ax[1].quiver(0, 0, eigval_A * eigvec_A[0, :], eigval_A * eigvec_A[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[1].text(-0.7, -0.4, '${p}_{1}$')
ax[1].text(-0.2, -0.8, '${p}_{2}$')
ax[1].set_xlim(-2, 1.5)
ax[1].set_ylim(-3, 0.5)
ax[1].set_xticks(np.arange(-2, 1.1, 1))
ax[1].set_yticks(np.arange(-3, 0.1, 1))
ax[1].set_title('after mapping')
ax[1].set(adjustable='box-forced', aspect='equal')
ax[1].grid()
plt.show()
```

上記から行列Aを作用させることでベクトル$p_1, p_2$は方向を変えず、長さのみ変わっていることが確認できる。

##### 固有値と固有ベクトルの意味

ベクトル$x_1, x_2$は$p_1, p_2$を用いて以下のように表せる
$$
{\bf x_1} = \left(\begin{array}{r} 1 \\ 0 \\ \end{array}\right) = -2 \sqrt{2}{\bf p_1} + \sqrt{5}{\bf p_2},
{\bf x_2} = \left(\begin{array}{r} 0 \\ 1 \\ \end{array}\right) = \sqrt{2}{\bf p_1} - \sqrt{5}{\bf p_2}
$$

```python
# 上記を可視化
x1 = np.array([[1], [0]])
x2 = np.array([[0], [1]])
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(8, 8))
coeff1 = np.array([-2 * np.sqrt(2), np.sqrt(5)]) # 固有ベクトルの係数
ax[0].quiver(x1[0, 0], x1[1, 0], color='red', angles='xy', scale_units='xy', scale=1)
ax[0].quiver(0, 0, coeff1 * eigvec_A[0, :], coeff1 * eigvec_A[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[0].text(1.0, -0.7, '${x}_{1}$')
ax[0].text(2.0, 1.6, '$-2\sqrt{2}{p}_{1}$')
ax[0].text(-3.5, -2.5, '$\sqrt{5}{p}_{2}$')
ax[0].set_xlim(-6, 6)
ax[0].set_ylim(-6, 6)
ax[0].set_xticks(np.arange(-6, 6.1, 2))
ax[0].set_yticks(np.arange(-6, 6.1, 2))
ax[0].set(adjustable='box-forced', aspect='equal')
ax[0].grid()
coeff2 = np.array([np.sqrt(2), -np.sqrt(5)]) # 固有ベクトルの係数
ax[1].quiver(x2[0, 0], x2[1, 0], color='blue', angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, coeff2 * eigvec_A[0, :], coeff2 * eigvec_A[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[1].text(-1.2, 0.5, '${x}_{1}$')
ax[1].text(-3.0, -2.0, '$\sqrt{2}{p}_{1}$')
ax[1].text(1.2, 1.5, '$-\sqrt{5}{p}_{2}$')
ax[1].set(adjustable='box-forced', aspect='equal')
ax[1].grid()
plt.show()
```

2つのベクトルを合成するとそれぞれ$x_1, x_2$になることが視覚的に確認できる。<br>
続いて、行列Aを写像と考えると、$x_1, x_2$の像は以下のようになる。

$$
{\bf A x_1} = -2 \sqrt{2}{\bf A p_1} + \sqrt{5}{\bf A p_2} = -2 \sqrt{2}2{\bf p_1} + \sqrt{5}3{\bf p_2}
$$

$$
{\bf A x_2} = \sqrt{2}{\bf A p_1} - \sqrt{5}{\bf A p_2} = \sqrt{2}2{\bf p_1} - \sqrt{5}3{\bf p_2}
$$

上記は$p_1$の方向に2倍、$p_2$の方向には3倍拡大される

```python
# 上記を可視化
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(8, 8))
# ベクトルx1の各成分の行列Aによる像
x1_comp_mapped = A.dot(coeff1 * eigvec_A)
ax[0].quiver(y1[0, 0], y1[1, 0], color='red', angles='xy', scale_units='xy', scale=1)
ax[0].quiver(0, 0, x1_comp_mapped[0, :], x1_comp_mapped[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[0].text(0.5, -0.8, '$A{x}_{1}$')
ax[0].text(-0.2, 3.5, '$-4\sqrt{2}{p}_{1}$')
ax[0].text(-5.3, -4.5, '$3\sqrt{5}{p}_{2}$')
ax[0].set_xlim(-6, 6)
ax[0].set_ylim(-6, 6)
ax[0].set_xticks(np.arange(-6, 6.1, 2))
ax[0].set_yticks(np.arange(-6, 6.1, 2))
ax[0].set(adjustable='box-forced', aspect='equal')
ax[0].grid()
# ベクトルx2の各成分の行列Aによる像
x2_comp_mapped = A.dot(coeff2 * eigvec_A)
ax[1].quiver(y2[0, 0], y2[1, 0], color='blue', angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, x2_comp_mapped[0, :], x2_comp_mapped[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[1].text(-1.5, 2.5, '$A{x}_{1}$')
ax[1].text(-3.5, -0.8, '$2\sqrt{2}{p}_{1}$')
ax[1].text(-1.8, 4.5, '$-3\sqrt{5}{p}_{2}$')
ax[1].set(adjustable='box-forced', aspect='equal')
ax[1].grid()
plt.show()
```

上記から行列Aにより、ベクトル$p_1, p_2$の方向は変化せず、固有値の分だけ長さが変わることがわかる。

#### 行列の階数（ランク）

上の例では、行列の固有値、固有ベクトルは2つ存在しAを掛けることで、回転し長さが変化したが、以下の行列のように当てはまらないケースもある<br>
$$
B = \left(
      \begin{array}{rr}
        1 & 2 \\
        2 & 4 \\
      \end{array}
    \right)
$$

ベクトル$x_1, x_2$のそれぞれに行列Bを掛けた式は以下。

$$
{\bf Bx_1} = \left(
               \begin{array}{rr}
                 1 & 2 \\
                 2 & 4 \\
               \end{array}
             \right)
             \left(
               \begin{array}{r}
                 1 \\
                 0 \\
               \end{array}
             \right)
=
             \left(
               \begin{array}{r}
                 1 \\
                 2 \\
               \end{array}
             \right)
$$

$$
{\bf Bx_2} = \left(
               \begin{array}{rr}
                 1 & 2 \\
                 2 & 4 \\
               \end{array}
             \right)
             \left(
               \begin{array}{r}
                 0 \\
                 1 \\
               \end{array}
             \right)
=
             \left(
               \begin{array}{r}
                 2 \\
                 4 \\
               \end{array}
             \right)
$$

```python
# ベクトルに行列を掛けた前後を可視化
x1 = np.array([[1], [0]])
x2 = np.array([[0], [1]])
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(8, 8))
ax[0].quiver(0, 0, np.hstack((x1, x2))[0, :], np.hstack((x1, x2))[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[0].text(0.7, -0.4, '${x}_{1}$')
ax[0].text(-0.5, 0.5, '${x}_{2}$')
ax[0].set_title('original')
ax[0].set_xlim(-1, 4.5)
ax[0].set_ylim(-1, 4.5)
ax[0].set_xticks(np.arange(-1, 4.1, 1))
ax[0].set_yticks(np.arange(-1, 4.1, 1))
ax[0].set(adjustable='box-forced', aspect='equal')
ax[0].grid()
ax[1].quiver(0, 0, [1, 2], [2, 4], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[1].text(0.8, 1.2, '$B{x}_{1}$')
ax[1].text(1.8, 3.2, '$B{x}_{2}$')
ax[1].set(adjustable='box-forced', aspect='equal')
ax[1].grid()
plt.show()
```

上記から行列Bをかけると、ベクトルの像は同じ方向を向くことが確認できる

```python
# 固有ベクトル、固有値を確認
B = np.array([[1, 2], [2, 4]])
eigval_B, eigvec_B = np.linalg.eig(B)
print(eigval_B)
print(eigvec_B)
```

```python
# 固有ベクトル、固有値を確認
B = np.array([[1, 2], [2, 4]])
eigval_B, eigvec_B = np.linalg.eigh(B)
print(eigval_B)
print(eigvec_B)
```

よって以下のようになる
$$
{\bf Bq_1} = 0 {\bf q_1}, {\bf Bq_2} = 5 {\bf q_2}
$$
固有ベクトルは数値計算上の誤差や表示桁数が有限であることを除けば、以下のように表せる
$$
{\bf q_1} = \frac{1}{\sqrt{5}} \left( \begin{array}{r} -2 \\ 1 \\ \end{array} \right), {\bf q_2} = \frac{1}{\sqrt{5}} \left( \begin{array}{r} 1 \\ 2 \\ \end{array} \right)
$$
ベクトル$x_1, x_2$を固有ベクトル$q_1, q_2$で表すと以下のようになる
$$
{\bf x_1} = \left( \begin{array}{r} 1 \\ 0 \\ \end{array} \right) = -\frac{2}{\sqrt{5}}{\bf q_1}+\frac{1}{\sqrt{5}}{\bf q_2},
{\bf x_2} = \left( \begin{array}{r} 0 \\ 1 \\ \end{array} \right) = \frac{1}{\sqrt{5}}{\bf q_1}+\frac{2}{\sqrt{5}}{\bf q_2}
$$

```python
# 上記を可視化
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(8, 8))
coeff1 = np.array([-2 / np.sqrt(5), 1 / np.sqrt(5)]) # 固有ベクトルの係数
ax[0].quiver(1, 0, color='red', angles='xy', scale_units='xy', scale=1)
ax[0].quiver(0, 0, coeff1 * eigvec_B[0, :], coeff1 * eigvec_B[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[0].text(1.0, -0.25, '${x}_{1}$')
ax[0].text(0.0, -0.8, '$-2/\sqrt{5}{q}_{1}$')
ax[0].text(-0.7, 0.4, '$1/\sqrt{5}{q}_{2}$')
ax[0].set_xlim(-1.5, 1.5)
ax[0].set_ylim(-1.5, 1.5)
ax[0].set(adjustable='box-forced', aspect='equal')
ax[0].grid()
coeff2 = np.array([1 / np.sqrt(5), 2 / np.sqrt(5)]) # 固有ベクトルの係数
ax[1].quiver(0, 1, color='blue', angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, coeff2 * eigvec_B[0, :], coeff2 * eigvec_B[1, :], color=['red', 'blue'], angles='xy', scale_units='xy', scale=1)
ax[1].text(-0.3, 0.8, '${x}_{1}$')
ax[1].text(-1.2, 0.3, '$1/\sqrt{5}{q}_{1}$')
ax[1].text(0.4, 0.8, '$2/\sqrt{5}{q}_{2}$')
ax[1].set(adjustable='box-forced', aspect='equal')
ax[1].grid()
plt.show()
```

上記の図から、$x_1, x_2$はそれぞれ2つのベクトルの合成で表されることが確認できる。<br>
$x_1, x_2$の行列Bによる像は以下のようになる。

$$
{\bf Bx_1} = -\frac{2}{\sqrt{5}}{\bf Bq_1} + \frac{1}{\sqrt{5}}{\bf Bq_2} = -\frac{2}{\sqrt{5}}0{\bf q_1} + \frac{1}{\sqrt{5}}5{\bf q_2} = \sqrt{5}{\bf q_2}
$$
$$
{\bf Bx_2} = \frac{1}{\sqrt{5}}{\bf Bq_1} + \frac{2}{\sqrt{5}}{\bf Bq_2} = \frac{2}{\sqrt{5}}0{\bf q_1} + \frac{2}{\sqrt{5}}5{\bf q_2} = 2\sqrt{5}{\bf q_2}
$$

以上の結果、行列Bには2つの固有ベクトルが存在するが、固有値0に対応する固有ベクトル$q_1$は潰れていることが確認できる。<br>
これは行列Bの固有ベクトルが独立ではなく、もうひとつのベクトルの倍数で表せるため。<br>
このとき行列の階数（ランク）は線形独立な列ベクトルの個数と定義時、行列Bの場合1となる。
