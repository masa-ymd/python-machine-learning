目的関数-> 最小化したいコスト関数J（連続値の活性化関数）

$$
\begin{eqnarray}
J(w) = \frac{1}{2}\sum_{i}(y^{(i)}-\phi(z^{(i)}))^2
\end{eqnarray}
$$

微分可能、凸関数-> 勾配降下法
勾配降下法をつかって重みを更新するには$$J(w)$$の勾配$$\nabla J(w)$$（重みベクトルwに関する偏導関数）に沿って1ステップ進む

$$
w := w + \Delta w
$$
$$\Delta w $$は負の勾配に学習率 $$\eta$$ を掛けたもの
$$
\Delta w = - \eta \nabla J(w)
$$

$$
\nabla J(w) = \frac{\partial J}{\partial w_{j}} 
$$
$$
= \frac{\partial}{\partial w_{j}} \frac {1}{2}\sum_{i}(y^{(i)} - \phi(x^{(i)}))^{2}
$$
$$
= \frac{1}{2}\frac{\partial}{\partial w_{j}} \sum_{i}(y^{(i)} - \phi(x^{(i)}))^{2}
$$
$$
= \frac{1}{2}\sum_i 2(y^{(i)} - \phi (z^{(i)}))\frac {\partial}{\partial w_j}(y^{(i)} - \phi(z^{(i)}))
$$
$$
= \sum_i(y^{(i)} - \phi(z^{(i)}))\frac{\partial}{\partial w_j}[y^{(i)} - \sum_k(w_k x_k^{(i)})]
$$
$$
= \sum_i(y^{(i)} - \phi(z^{(i)}))(-x_j^{(i)})
$$
$$
= - \sum_i(y^{(i)} - \phi(z^{(i)}))x_j^{(i)}
$$
$$
\Delta w_j = - \eta \frac{\partial J}{\partial w_j} = \eta \sum_i(y^{(i)} - \phi(z^{(i)}))x_j^{(i)}
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
# 1-100行目の目的変数抽出
y = df.iloc[0:100, 4].values
# Iris-setosaを-1に、Iris-versicolorを1に変換
y = np.where(y == 'Iris-setosa', -1, 1)
# 1-100行目の1, 3項目目を抽出
X = df.iloc[0:100, [0,2]].values
```

```python
class AdalineGD(object):
    """
    ADAptive LInear NEuron 分類器
    
    パラメータ
    ----------
    eta : float
        学習率（0.0 <= eta <= 1.0)
    n_iter : int
        トレーニング回数
        
    属性
    ----------
    w_ : 一次元配列
       重み
    errors_ : リスト
       各トレーニングでの誤分類数
    """
    
    def __init__(self, eta=0.01, n_iter=50):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self, X, y):
        """
        トレーニングデータに適合させる
        
        パラメータ
        ----------
        X : {サンプル数, 特徴量の個数} 
          トレーニングデータ
        y : {サンプル数}
          目的変数
          
        戻り値
        ----------
        self : object
        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []
        
        for i in range(self.n_iter): # n_iter回反復してトレーニング
            # 活性化関数の出力の計算 phi(w^Tx) = w^Tx
            output = self.net_input(X)
            # 誤差 y^(i) - phi(w^Tx)
            errors = (y - output)
            # w_1…mの更新（eta * errors . X)
            self.w_[1:] += self.eta * X.T.dot(errors)
            # w_0 の更新
            self.w_[0] += self.eta * errors.sum()
            # コスト関数の計算
            cost = (errors**2).sum() / 2.0
            # コストを保存
            self.cost_.append(cost)
        return self
    
    def net_input(self, X):
        """総入力を計算"""
        return np.dot(X, self.w_[1:]) + self.w_[0] * 1
    
    def activation(self, X):
        """線形活性化関数の出力を計算"""
        return self.net_input(X)
    
    def predict(self, X):
        """1ステップ後のクラスラベルを返す"""
        return np.where(self.activation(X) >= 0.0, 1, -1)
```

```python
# 描画領域を1行2列に分割
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
# 勾配降下法による学習(eta=0.01)
ada1 = AdalineGD(n_iter=10, eta=0.01).fit(X, y)
# トレーニング回数とコスト（対数表示）のプロット
ax[0].plot(range(1, len(ada1.cost_) + 1), np.log10(ada1.cost_), marker='o')
# 軸のラベル設定
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log(Sum-squared-error)')
# タイトルの設定
ax[0].set_title('Adaline - Learning rate 0.01')
# 勾配降下法による学習(eta=0.0001)
ada2 = AdalineGD(n_iter=10, eta=0.0001).fit(X, y)
# トレーニング回数とコスト（対数表示）のプロット
ax[1].plot(range(1, len(ada2.cost_) + 1), np.log10(ada2.cost_), marker='o')
# 軸のラベル設定
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('log(Sum-squared-error)')
# タイトルの設定
ax[1].set_title('Adaline - Learning rate 0.0001')
# 図の表示
plt.show()
```

スケーリング手法　標準化（standardization）

$$
\bf{x'_j} = \frac {\bf{x_j} - \bf{\mu_j}} {\bf{\sigma_j}}
$$

```python
# データコピー
X_std = np.copy(X)
# 各列の標準化
X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02, xlabel='x', ylabel='y', legend_loc='upper left'):
    
    # マーカーとカラーマップの準備
    markers = ('s', 'o', 'x', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # 決定領域のプロット
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    # グリッドポイントの生成
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    print 'xx1'
    print xx1
    print 'xx2'
    print xx2
    print 'xx1.ravel()'
    print xx1.ravel()
    print 'xx2.ravel()'
    print xx2.ravel()
    
    # 各特徴量を1次元配列に変換して予測を実行
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    # 予測結果を元のグリッドポイントのデータサイズに変換
    Z = Z.reshape(xx1.shape)
    print 'Z'
    print Z
    # グリッドポイントの等高線のプロット
    plt.contourf(xx1, xx2, Z, aplha=0.4, cmap=cmap)
    # 軸の範囲の設定
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # クラスごとにサンプルをプロット
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=cmap(idx), marker=markers[idx], label=cl)
        
    # 軸のラベルの設定
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # 凡例の設定
    plt.legend(loc=legend_loc)
    # 図の表示
    plt.show()
```

```python
# 学習（eta=0.01）
ada = AdalineGD(n_iter=15, eta=0.01)
# モデルの適合
ada.fit(X_std, y)
# 決定境界のプロット
plot_decision_regions(X_std, y, classifier=ada, xlabel='sepal length [standardized]', ylabel='petal length [standardized]')

# トレーニング回数とコスト（対数表示）のプロット
plt.plot(range(1, len(ada.cost_) + 1), np.log10(ada.cost_), marker='o')
# 軸のラベル設定
plt.xlabel('Epochs')
plt.ylabel('log(Sum-squared-error)')
# タイトルの設定
plt.title('Adaline - Learning rate 0.01')
# 図の表示
plt.show()
```
