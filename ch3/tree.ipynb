# 決定木学習

情報利得（分割後の集合のばらつきの減少）が最大になる特徴量でデータを分割していく<br>
情報利得（目的関数）は以下のように定義される
$$
IG(D_p, f) = I(D_p) - \sum_{j=1}^m \frac{N_j}{N_p}I(D_j)
$$

$D_p$は親のデータセット、$D_j$はj番目の子ノードのデータセット、$I$は不純度（異なるサンプルがどの程度混じっているか）、$N_p$は親ノードのサンプルの総数、$N_j$はj番目の子ノードのサンプルの個数<br>
シンプルな2分決定木の場合は以下のように定義する<br>
$$
IG(D_p, f) = I(D_p) - \frac{N_{left}}{N_p}I(D_{left}) - \frac{N_{right}}{N_p}I(D_{right})
$$

２分決定木でよく使用される不純度の指標は、ジニ不純度、エントロピー、分割誤差の３種類がある

### エントロピー

$$
I_H(t) = - \sum_{i=1}^c p(i|t)log_2p(i|t)
$$

$p(i|t)$は特定のノードtにおいてクラスiに属するサンプルの割合を示す<br>
２値分類でエントロピーが最大になるのは、クラスが$p(i=1|t)=0.5$および$p(i=0|t)=0.5$で一様に分布している時

### ジニ不純度

$$
I_G(t) = \sum_{i=1}^cp(i|t)(1 - p(i|t)) = 1 - \sum_{i=1}^cp(i|t)^2
$$

エントロピーと同様に最大になるのはクラスが完全に混合されている場合

### 分類誤差

$$
I_E{t} = 1 - max\{p(i|t)\}
$$

これは木の剪定に役立つがノードのクラス確率の変化にあまり敏感でないため、決定木を成長させるのにあまり適していない

### 情報利得を計算する例

クラス1:クラス2=40:40の親ノードを、クラス1:クラス2=30:10の子ノード1と、クラス1:クラス2=10:30の子ノード2に分割した場合の
ジニ不純度を計算してみる。

親ノードの不純度
$$
I_G(D_p) = 1 - \Biggl(\Biggl(\frac{40}{80}\Biggl)^2 + \Biggl(\frac{40}{80}\Biggl)^2\Biggl) = 0.5
$$
子ノード1の不純度
$$
I_G(D_c1) = 1 - \Biggl(\Biggl(\frac{30}{40}\Biggl)^2 + \Biggl(\frac{10}{40}\Biggl)^2\Biggl) = 0.375
$$
子ノード2の不純度
$$
I_G(D_c2) = 1 - \Biggl(\Biggl(\frac{10}{40}\Biggl)^2 + \Biggl(\frac{10}{30}\Biggl)^2\Biggl) = 0.375
$$
情報利得
$$
IG_G = 0.5 - \frac{40}{80}\times0.375 - \frac{40}{80}\times0.375 = 0.125
$$

## 3種類の不純度条件を視覚的に比較する

```python
import matplotlib.pyplot as plt
import numpy as np

# ジニ不純度
def gini(p):
    return (p) * (1 - (p)) + (1 - p) * (1 - (1 - p))

# エントロピー
def entropy(p):
    return - p * np.log2(p) - (1 - p) * np.log2((1 - p))

# 分類誤差
def error(p):
    return 1 - np.max([p, 1 - p])

# 確率を表す配列
x = np.arange(0.0, 1.0, 0.01)

# エントロピーを計算
ent = [entropy(p) if p != 0 else None for p in x]
sc_ent = [e * 0.5 if e else None for e in ent]
# 分類誤差
err = [error(i) for i in x]

# 図を作成
fig = plt.figure()
ax = plt.subplot(111)
# それぞれの不純度をループ処理でプロット
for i, lab, ls, c, in zip([ent, sc_ent, gini(x), err],
                          ['Entropy', 'Entropy(scaled)', 'Gini Impurity', 'Misclassification Error'],
                          ['-', '-', '--', '-.'],
                          ['black', 'lightgray', 'red', 'green', 'cyan']):
    ax.plot(x, i, label=lab, linestyle=ls, lw=2, color=c)

# 凡例
ax.legend(loc='upper centor', bbox_to_anchor=(1.2, 1.2), ncol=3, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```

## 決定木の構築

```python
from sklearn import datasets
import numpy as np

# iris データセットをロード
iris = datasets.load_iris()
# 3,4列目の特徴量を抽出
X = iris.data[:, [2, 3]]
# クラスラベルを取得
y = iris.target

from sklearn.model_selection import train_test_split
# 全体の30% をテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# トレーニングデータの平均と標準偏差を計算
sc.fit(X_train)
# 計算した結果を元にデータを標準化
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

# トレーニングデータとテストデータを行方向に結合
X_combined_std = np.vstack((X_train_std, X_test_std))
# 同様にクラスラベルも結合
y_combined = np.hstack((y_train, y_test))
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02, xlabel='x', ylabel='y', title='title', legend_loc='upper left'):
    
    # マーカーとカラーマップの準備
    markers = ('s', 'o', 'v', '^', 'x')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # 決定領域のプロット
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    # グリッドポイントの生成
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    # 各特徴量を1次元配列に変換して予測を実行
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    # 予測結果を元のグリッドポイントのデータサイズに変換
    Z = Z.reshape(xx1.shape)
    # グリッドポイントの等高線のプロット
    plt.contourf(xx1, xx2, Z, aplha=0.4, cmap=cmap)
    # 軸の範囲の設定
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # クラスごとにサンプルをプロット
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=cmap(idx), marker=markers[idx], label=cl)
        
    # テストサンプルを目立たせる
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], c='',
                   alpha=1.0, linewidths=1, marker='o',
                   s=55, label='test set')
        
    # 軸のラベルの設定
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # タイトル設定
    plt.title(title)
    # 凡例の設定
    plt.legend(loc=legend_loc)
    # 図の表示
    plt.show()
```

決定木学習では特徴空間を矩形に分割することで複雑な決定境界を構築できる。<br>
ただし、木が深くなるほど境界が複雑になり、過学習に陥りやすくなる。<br>
特徴量のスケーリングはしきい値の値が変化するだけで、結果には影響しない。<br>

```python
from sklearn.tree import DecisionTreeClassifier
# エントロピーを指標とする決定木のインスタンスを生成
tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
# モデルにデータをfitさせる
tree.fit(X_train, y_train)
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X_combined, y_combined, classifier=tree, test_idx=range(105, 150),
                      xlabel='petal length [cm]', ylabel='petal width [cm]')
```

```python
# 結果の樹形図を出力する
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file='tree.dot', feature_names=['petal length', 'petal width'])
```

## ランダムフォレスト
弱い学習アルゴリズムを組み合わせ、より頑健なモデルを作る。<br>
頑健なモデルは、汎化誤差が改善され、過学習に陥りにくい<br>
ランダムフォレストは以下の4つのステップで実行される

1. トレーニングデータセットからn個のサンプルをランダムに（復元）選択
2. 上記のサンプルで決定木を成長させる
   * d個の特徴量をランダムに（非復元）抽出する
   * 目的関数にしたがってノードを分割する
3. ステップ1〜2をk回繰り返す
4. 決定木毎の予測をまとめ、多数決に基づいてクラスラベルを割り当てる

### ランダムフォレストの特徴

* 決定木と同じレベルの意味解釈可能性はない
* アンサンブルモデルは個々の決定木のノイズにかなり強く剪定の必要がない
* 考慮すべきパラメータはステップ3のk kを多くすれば性能が上がるが、計算コストも増える
* サンプル数nと特徴量の数dは最適化が可能だが、あまり一般的ではない
* nを大きくするとランダム性が低下し、過学習に陥る可能性が上がる 殆どの実装ではnは元のデータセットと同じ大きさになるようサンプルを選択する
* dの妥当なデフォルト値はmが特徴量の数とした場合、$d = \sqrt{m}$

```python
from sklearn.ensemble import RandomForestClassifier
# エントロピーを指標とするランダムフォレストのインスタンスを生成
forest = RandomForestClassifier(criterion='entropy', n_estimators=10, random_state=1, n_jobs=2)
# モデルにデータをfitさせる
forest.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined, classifier=forest, test_idx=range(105, 150),
                      xlabel='petal length [cm]', ylabel='petal width [cm]')
```
