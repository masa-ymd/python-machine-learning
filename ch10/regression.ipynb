### 単純な線形回帰モデル

線形回帰の目標は説明変数と目的変数の関係を表す式の重みwを学習すること<br>
簡単な表現に変えると、サンプル点を通過する直線のうち、最も適合するものを見つけ出すこと。この線は回帰直線と呼ばれる<br>
回帰直線とサンプル点との距離はオフセット、または残差と呼ばれ、予測誤差を表す<br>
一般化した線形モデルの式（重回帰）は以下
$$
y = w_0x_0 + w_1x_1 + \cdots + w_mx_m = \sum_{i=0}^m w_ix_i = \bf w^Tx 
$$
$w_0$はy軸の切片で$x_0 = 1$<br>

#### 相関行列

相関行列はピアソンの積率相関係数を成分とする正方行列で、その成分は特徴量のペアの線形的な従属関係を数値化するもので、<br>
よくピアソンのrと略される。<br>
特徴量xとyの相関係数rは以下のように、2つの特徴量の共分散をそれらの標準偏差の積で割ったものとして表される

$$
r = \frac{\sum_{i=1}^n \left[ (x^{(i)} - \mu_x)(y^{(i)} - \mu_y) \right]}{\sqrt{\sum_{i=1}^n(x^{(i)} - \mu_x)^2}\sqrt{\sum_{i=1}^n(y^{(i)} - \mu_y)^2}} = \frac{\sigma_{xy}}{\sigma_x \sigma_y}
$$
$\mu$は対応する特徴量の標本平均で$\sigma_xy$は特徴量xyの間の共分散、$\sigma_x$、$\sigma_y$はそれぞれの特徴量の標準偏差<br>
特徴は以下

* 取る値は-1から1まで
* r=1の場合、2つの特徴量は完全な正の相関がある
* r=0の場合は相関がない
* r=-1の場合は完全に負の相関がある

相関行列は標準化されたデータから計算した共分散行列に等しい<br>

#### 最小2乗回帰モデル

最小2乗法（OLS）を用い、サンプル点と予測点の距離（残差）の合計を最小にするパラメータを推定する<br>
ADALINEのコスト関数とOLSのコスト関数は等しく、$\hat{y} = \bf w^Tx$（予測された値）とすると以下のように表せる
$$
J(w) = \frac{1}{2}\sum_{i=1}^n(y^{(i)} - \hat{y}^{(i)})^2
$$
基本的にはOLS線形回帰は単位ステップ関数がないADALINEとして解釈できる

```python
# OLSを実装
class LinearRegressionGD(object):
    
    # 初期化
    def __init__(self, eta=0.001, n_iter=20):
        self.eta = eta       # 学習率
        self.n_iter = n_iter # トレーニング回数
        
    # トレーニング
    def fit(self, X, y):
        self.w_ = np.zeros(1 + X.shape[1])            # 重み（特徴量=列の数分）を初期化 +1 はw_0の分
        self.cost_ = []                               # コスト関数の値を初期化
        for i in range(self.n_iter):
            output = self.net_input(X)                # 活性化関数の出力を計算
            errors = (y - output)                     # 誤差を計算
            self.w_[1:] += self.eta * X.T.dot(errors) # 重みw_1以降を更新 (P34 2.5.5 参照)
            self.w_[0] += self.eta * errors.sum()     # 重みw_0を更新 x_0=1で考えると誤差の合計
            cost = (errors ** 2).sum() / 2.0          # コスト関数を計算
            self.cost_.append(cost)                   # コスト関数の結果を配列に格納
        return self
    
    # 総入力を計算
    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    # 予測値を計算
    def predict(self, X):
        return self.net_input(X)
```

```python
# MEDV-RMでトレーニング
X = df[['RM']].values
y = df[['MEDV']].values.ravel()
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_std = sc_X.fit_transform(X)
y_std = sc_y.fit_transform(y)
lr = LinearRegressionGD()
lr.fit(X_std, y_std)
```

```python
# トレーニング回数とコストの関係をプロット
plt.plot(range(1, lr.n_iter+1), lr.cost_)
plt.ylabel('SSE')
plt.xlabel('Epoch')
plt.show()
```

```python
# 散布図と回帰直線を描画する簡単な関数を定義
def lin_regplot(X, y, model, xlabel='X', ylabel='y',):
    plt.scatter(X, y, c='blue')
    plt.plot(X, model.predict(X), color='red')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    return None
```

```python
lin_regplot(X_std, y_std, lr, xlabel='Average number of room [RM] (standardized)', ylabel='Price in $1000\'s [MEDV] (standerdized)')
plt.show()
```

上記の直線から、部屋の数が増えると住宅価格も上昇するという傾向が分かる<br>
y=3の箇所に直線が存在するが、これは住宅価格がそこで打ち切られている可能性を示している

```python
# 予測結果を元の尺度に戻す
num_room_std = sc_X.transform([5.0])
price_std = lr.predict(num_room_std)
print("Price in $1000's: %.3f" % sc_y.inverse_transform(price_std))
```

### RANSACを使ったロバストな回帰モデル

線形回帰モデルは外れ値に大きく左右され、場合によっては一部の外れ値がモデルの係数の推定に大きな影響を与えることがある<br>
統計学的検定法により、外れ値を検出し除去する方法もあるが、それに変わる方法として、ロバスト回帰がある<br>
ロバスト回帰は外れ値の影響を抑えて回帰を行う手法であり、基本的なアイデアは外れ値の重みを小さくすることにある<br>
ロバスト回帰の１つであるRANSACの手順は以下<br>

1. 求めるパラメータの数に合わせて、全てのデータからn個のデータをピックアップする
2. ピックアップしたデータで、パラメータを求める
3. 上記のパラメータのモデルで、ピックアップしたデータ以外のデータに対して推定値を求め、誤差を算出する
4. 誤差が（予め決めた）許容範囲であれば、モデル候補に加える
5. 1〜4を繰り返し、一番精度の良いパラメータを選択する
6. 一番よいパラメータを用い、すべてのデータに再度式を適用し、誤差が許容範囲内のデータを抽出する
  （この時点で雑音データがほぼ取り除かれる）
7. 抽出したデータを元に、再度パラメータを求める
8. 求まったパラメータがこのデータのもっともらしいパラメータとなる

```python
# scikit-learnのRANSACを利用する
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),
                         max_trials=100,  # イテレーションの回数
                         min_samples=50,  # ランダムに選択されるサンプルの最小数
                         residual_metric=lambda x: np.sum(np.abs(x), axis=1), # 残差の基準は各サンプルの特徴量の絶対値の和
                         residual_threshold=5.0, # 学習曲線に対する距離の閾値（この場合は5単位以内を正常値にする）
                         random_state=0
                        )
ransac.fit(X, y)
```

```python
# 正常値と外れ値、回帰直線を描画
inlier_mask = ransac.inlier_mask_                     # 正常値を表す真偽値を取得
outlier_mask = np.logical_not(inlier_mask)            # 外れ値を表す真偽値を取得
line_X = np.arange(3, 10, 1)                          # X軸の値
line_y_ransac = ransac.predict(line_X[:, np.newaxis]) # 予測値を計算
# 正常値をプロット
plt.scatter(X[inlier_mask], y[inlier_mask], c='blue', marker='o', label='Inliers')
# 外れ値をプロット
plt.scatter(X[outlier_mask], y[outlier_mask], c='lightgreen', marker='s', label='Outliers')
# 予測値をプロット
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in $1000\'s [MEDV]')
plt.legend(loc='upper left')
plt.show()
```

線形回帰モデルは正常値（青丸）の範囲に適合している

```python
# モデルの切片と傾きを表示する
print('Slope: %.3f' % ransac.estimator_.coef_[0])
print('Intercept: %.3f' % ransac.estimator_.intercept_)
```

残差プロットは回帰モデルを診断して、非線形性や外れ値を検出し、誤差がランダムに分布しているかを調べるために利用できる<br>
線形回帰モデルでは残差について以下の3つの仮定が置かれており、これをチェックすることは定石となっている<br>

2. 互いに無相関

#### 平均2乗誤差（MSE）

モデルの性能を数値化する方法の１つで以下の式で表せる<br>
$$
MSE = \frac{1}{n}\sum_{i=1}^n(y^{(i)} - \hat{y}^{(i)})^2
$$

```python
from sklearn.metrics import mean_squared_error
# MSE出力
print('MSE train: %.3f, test: %.3f' % (mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred)))
```

上記はトレーニングデータとテストデータの差があり、過学習が発生していることを示す

#### 決定係数（$R^2$）

決定係数は以下の式で表される<br>

$$
R^2 = 1 - \frac{SSE}{SST} = 1 - \frac{MSE}{Var(y)}
$$

分子のSSEは回帰モデルで説明できない目的変数のばらつきを表し、分母のSSTは目的変数の平均値のまわりにおけるばらつきを表している<br>
したがって、$R^2$は目的変数の変動の内、回帰モデルによって説明できる割合を表している

```python
from sklearn.metrics import r2_score
# R^2出力
print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred)))
```

上記も同様にトレーニングデータとテストデータの差があり、過学習が発生していることを表している

### 回帰モデルでの正則化手法

* リッジ回帰
$$
J(w)_{Ridge} = \sum_{i=1}^n(y^{(i)} - \hat{y}^{(i)})^2 + \lambda \sum_{j=1}^m w_j^2
$$
$\lambda$の値を大きくすると正則化の強さが上がり、重みが小さくなる
* LASSO
$$
J(w)_{LASSO} = \sum_{i=1}^n(y^{(i)} - \hat{y}^{(i)})^2 + \lambda \sum_{j=1}^m |w_j|
$$
LASSOは正則化の強さによっては特定の変数の重みが0になることもあり、特徴選択の手法としても役立つ<br>
ただし、制約があり、サンプル数が特徴量よりも少ない場合、特徴量はサンプル数以下にはできない
* ElasticNet
$$
J(w)_{ElasticNet} = \sum_{i=1}^n(y^{(i)} - \hat{y}^{(i)})^2 + \lambda \sum_{j=1}^m w_j^2 + \lambda \sum_{j=1}^m |w_j|
$$
上記２つの折衷案で、$\lambda \sum_{j=1}^m |w_j|$で特徴量を減らしつつ、$\lambda \sum_{j=1}^m w_j^2$でLASSOの制約を部分的に克服する

```python
# 各モデルの初期化は以下のようにして実行可能
# リッジ回帰
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)  # alphaはペナルティ項の影響度合いを表す
# LASSO
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1.0)
# ElasticNet
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet(alpha=1.0, l1_ratio=0.5) # l1_ratioはLASSOのペナルティ項。この例では残りの0.5がl2(リッジ回帰)ペナルティ項に割り振られる
```

### 多項式回帰

説明変数と目的変数の関係が線形であるという前提が外れる場合、多項式モデルを使用する方法がある<br>
多項式回帰モデルの式は以下
$$
y = w_0 + w_1 x + w_2 x^2 + \cdots + w_d x^d
$$
dは多項式の次数<br>
多項式回帰により非線型関係をモデリングすることは可能だが、回帰の係数wは線形なので、重回帰モデルとみなされる

```python
# 線形回帰と多項式回帰を比較する
from sklearn.preprocessing import PolynomialFeatures
X = np.array([258.0, 270.0, 294.0, 320.0, 342.0, 368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2, 342.2, 360.8, 368.0, 391.2, 390.8])

# 両方のモデルをインスタンス化
lr = LinearRegression()
pr = LinearRegression()

# 多項式特徴量のクラスをインスタンス化し、データを変換
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)

# 学習（fit）
lr.fit(X, y)
pr.fit(X_quad, y)

# グラフ描画用にX軸のデータを作成
X_fit = np.arange(250, 600, 10)[:, np.newaxis] # newaxisで列ベクトルにする

# 上記のX軸データを対象に両方のモデルで予測値を出す
y_lin_fit = lr.predict(X_fit)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))

# グラフ作成
plt.scatter(X, y, label='training points')
plt.plot(X_fit, y_lin_fit, label='linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit, label='quadratic fit')
plt.legend(loc='upper left')
plt.show()
```

```python
# 数値的にモデルの性能を比較する
y_lin_pred = lr.predict(X)
y_quad_pred = pr.predict(X_quad)
print('Training MSE linear: %.3f, quadratic: %.3f' % (mean_squared_error(y, y_lin_pred), mean_squared_error(y, y_quad_pred)))
print('Training R^2 linear: %.3f, quadratic: %.3f' % (r2_score(y, y_lin_pred), r2_score(y, y_quad_pred)))
```

上記のようにグラフは多項式回帰のほうがデータをうまく補足しており、平均２乗誤差は大幅に改善、決定係数も多項式のほうが高くなっている

```python
# Housingデータセットを2次、3次多項式に適用する
# データセットと線形回帰のインスタンスを生成
X = df[['LSTAT']].values  # 低所得者の割合
y = df['MEDV'].values     # 住宅価格の中央値
regr = LinearRegression()

# 多項式用の特徴量を作成
quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)

# 線形回帰モデルの学習
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]
regr = regr.fit(X, y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y, regr.predict(X))

# 2次の多項式モデルの学習
regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))

# 3次の多項式モデルの学習
regr = regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))

# 各モデルの結果をプロット
plt.scatter(X, y, label='training points', color='lightgreen')
plt.plot(X_fit, y_lin_fit, label='linear (d=1), $R^2=%.2f$' % linear_r2, color='blue', lw='2', linestyle=':')
plt.plot(X_fit, y_quad_fit, label='quadratic (d=2), $R^2=%.2f$' % quadratic_r2, color='red', lw='2', linestyle='-')
plt.plot(X_fit, y_cubic_fit, label='cubic (d=3), $R^2=%.2f$' % cubic_r2, color='green', lw='2', linestyle='--')
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000\'s [MEDV]')
plt.legend(loc='upper right')
plt.show()
```

3次の多項式がMEDVとLSTATの関係をもっともよく補足しているが、多項式特徴量を追加するほどモデルの複雑さが増し、過学習のリスクが高まる<br>
実際にはテストデータを使って、汎化性能を評価する必要がある<br>
更に、非線形関係のモデリングにおいて多項式の特徴量は最良の選択ではない場合もある<br>
例えば、変数を対数や平方根に変換することで、線形回帰モデルの学習に適した特徴空間にデータを射影するなどがあげられる<br>

```python
X_log = np.log(X)   # Xを対数変換
y_sqrt = np.sqrt(y) # yを平方根に変換

# グラフ描画用のX軸の値を生成
X_fit = np.arange(X_log.min() - 1, X_log.max() + 1, 1)[:, np.newaxis]

# 学習
regr = regr.fit(X_log, y_sqrt)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y_sqrt, regr.predict(X_log))

# グラフをプロット
plt.scatter(X_log, y_sqrt, label='training points', color='lightgray')
plt.plot(X_fit, y_lin_fit, label='linear (d=1), $R^2=%.2f$' % linear_r2, color='blue', lw=2)
plt.xlabel('log(% lower status of the population [LSTAT])')
plt.ylabel('$\sqrt{price \; in \; \$1000\'s [MEDV]}$')
plt.legend(loc='lower left')
plt.show()
```

説明変数を対数に、目的変数を平方根に変換すると２つの変数の関係をうまく補足できるようになり、最も良い決定値スコアが得られる

### 決定木を用いた非線形関係への対処

決定木…区分線形関数（区分によって異なる線形関数）の和として考えることができる<br>
決定木アルゴリズムにより、区分ごとに異なる分離基準を設けて、分離を行う（階段みたいな）<br>
決定木の利点としては特徴量の変換を行う必要がないことが挙げられる<br>
木を分割する基準は以下に示す情報利得(IG)を最大にできるかどうかである
$$
IG(D_p, x) = I(D_p) - \frac{1}{N_p}I
$$
* $D_p$は親ノードのトレーニングサンプルのサブセット
* xは分割の対象となる特徴量
* Iは不純度関数
* $N_p$は親ノードのサンプルの個数
回帰の場合は不純度Iをエントロピーではなく、以下の平均２乗誤差に置き換える
$$
I(t) = MSE(t) = \frac{1}{N_t} \sum_{i \in D_t}(y^{(i)} - \hat{y}^{(i)})^2
$$
* tはノードを表す
* $D_t$はノードtのトレーニングサンプル
* $N_t$はノードtのサンプル数
* $\hat{y} = \frac{1}{N} \sum_{i \in D_t}y^{(i)}$

```python
# 決定木の学習曲線をプロット
from sklearn.tree import DecisionTreeRegressor
X = df[['LSTAT']].values
y = df['MEDV'].values
tree = DecisionTreeRegressor(max_depth=3) # 深さを3に指定
tree.fit(X, y)
sort_idx = X.flatten().argsort() # 線グラフ描画のため、Xが昇順になるインデックスを作成
lin_regplot(X[sort_idx], y[sort_idx], tree, xlabel='% lower status of the population [LSTAT]', ylabel='Price in $1000\'s [MEDV]')
plt.show()
```

上記のように決定木はデータの一般的な傾向を補足している<br>
しかし、予測値は連続ではないため微分できない（勾配降下法のようなチューニングができない）<br>
またハイパーパラメータとして決定木の深さがあるが、適切な値を選択する必要がある<br>
以下のランダムフォレスト回帰の場合、アンサンブル学習のため汎化性能が高い<br>
またデータの外れ値に影響を受けず、パラメータのチューニングをそれほど必要としない（一般的には決定木の個数のみ）という利点もある<br>

```python
# Housingデータの全てを利用して、ランダムフォレスト回帰を実行
X = df.iloc[:, :-1].values
y = df['MEDV'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

from sklearn.ensemble import RandomForestRegressor
# criterionで決定木の成長条件となる指標を指定
forest = RandomForestRegressor(n_estimators=1000, criterion='mse', random_state=1, n_jobs=-1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)
print('MSE train: %.3f, test: %.3f' % (mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred)))
```

MSEの結果からランダムフォレスト回帰は過学習することが分かる

```python
# 残差プロット
plt.scatter(y_train_pred, y_train_pred - y_train, c='black', marker='o', s=35, alpha=0.5, label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='lightgreen', marker='s', s=35, alpha=0.7, label='Test data')
plt.xlabel('Predicted value')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10, 50])
plt.show()
```

残差は原点の中心に完全にランダムにプロットされているわけではないようなので、まだモデルが情報を補足できていないことを示唆している<br>
しかし、線形モデルの残差プロットと比較すると大きく改善しているように見える
