# 有益な特徴量の選択

テストデータよりもトレーニングデータセットの方がパフォーマンスがはるかに良い場合、過学習に陥っている<br>
モデルがトレーニングセットの観測結果にパラメータを適合させすぎ、複雑になっている、バリアンスが高い状態<br>
一般的な対処方法は以下の通り
* 更に多くのトレーニングデータを集める（難易度が高い）
* 正則化を通じて複雑さにペナルティを科す
* パラメータの数が少ない、より単純なモデルを選択する
* データの次元の数を減らす

### L1正則化による疎な解

L1正則化は以下の式で表される。

$$
L1: \|w\|_1 = \sum_{j=1}^m|w_j|
$$

L1正則化によって返されるのは疎な特徴ベクトルになり、ほとんどの特徴量の重みは0となる。<br>
特に無関係な次元の数がサンプルよりも多い場合に特徴量を選択する手法として有効な場合がある<br>
正則化は最小化の等高線と正則化の線がふれあう点が最善の値になる。

```python
# Wineデータセットを読み込み
import pandas as pd
import numpy as np
df_wine = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
    header=None)
# 列名を設定
df_wine.columns = [
    'Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoides', 'Nonflavanoid phenols',
    'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 ofdiluted wines', 'Proline']
# train_test_split関数でデータセットをランダムに分割する
from sklearn.model_selection import train_test_split
# 特徴量とラベルを別々に抽出
X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values
# 全体の30%をテストデータにして分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
# インスタンス生成
stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
# L1正則化
from sklearn.linear_model import LogisticRegression
# インスタンス生成
lr = LogisticRegression(penalty='l1', C=0.1)
# fit
lr.fit(X_train_std, y_train)
# トレーニングデータに対する正解率の表示
print('Training accuracy:', lr.score(X_train_std, y_train))
# テストデータに対する正解率の表示
print('Test accuracy:', lr.score(X_test_std, y_test))
```

```python
# 3クラスそれぞれを判定するための切片の表示
# 最初の項はクラス1対クラス2+3を判定する
lr.intercept_
```

```python
# 3クラスそれぞれを判定するためのモデルの重み係数の表示
lr.coef_
```

上記の切片、重みは以下のような式で表される
$$
z = w_1x_1 + \cdots + w_mx_m + b = \sum_{j=1}^mx_jw_j + b = \bf{w^Tx} + b
$$

```python
import matplotlib.pyplot as plt
# 描画の準備
fig = plt.figure()
ax = plt.subplot(111)
# 各係数の色リスト
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black',
         'pink', 'lightgreen', 'lightblue', 'gray', 'indigo', 'orange']
# 空のリストを生成
weights, params = [], []
# 逆正則化パラメータの値ごとに処理
for c in np.arange(-4, 6):
    lr = LogisticRegression(penalty='l1', C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    # クラス2判定用の重み(2番め)
    weights.append(lr.coef_[1])
    params.append(10**c)

# weightsをNumpy配列に変換
weights = np.array(weights)
# 各重み係数をプロット
for column, color in zip(range(weights.shape[1]), colors):
    # 横軸をC, 縦軸を重み係数としたグラフ
    plt.plot(params, weights[:, column], label=df_wine.columns[column + 1], color=color)
    
# y=0に黒い波線を引く
plt.axhline(0, color='black', linestyle='--', linewidth=3)
# 横軸の範囲設定
plt.xlim([10**(-5), 10**5])
# 軸のラベル設定
plt.ylabel('weight coefficient')
plt.xlabel('C')
# 横軸を対数スケールに設定
plt.xscale('log')
plt.legend(loc='upper left')
ax.legend(loc='upper center', bbox_to_anchor=(1.38, 1.03), ncol=1, fancybox=True)
plt.show()
```

Cの値を大きくする（正則化によるペナルティを強める）と各重み係数は0に近づく

## 逐次特徴選択アルゴリズム

次元削減は過学習を回避する方法の一つ

1. 特徴選択…元の特徴量の一部を選択する
2. 特徴抽出…新しい特徴部分空間を生成する

逐次特徴選択のアルゴリズムは貪欲探索。計算効率の改善、モデルの汎化誤差の削減が目的で利用される。

### 逐次後退選択（SBS）

アルゴリズムは以下の通り

1. k=dで初期化する。dは全体の特徴空間$\bf{X_d}$の次元数
2. 評価関数Jを最大化する特徴量$\bf{x^-}$を決定する。$\bf{x}$は$\bf{x} \in \bf{X_d}$。<br>
   $\bf{x^-} = argmaxJ(\bf{X_d} - \bf{x})$ …全体からxを引いた残りの$\bf{x^-}$が最も性能に影響がない
3. 特徴量の集合から特徴量$\bf{x^-}$を削除する<br>
   $\bf{X_d} := x; k := k -1$
4. kが目的の個数と等しければ終了、そうでない場合は2に戻る

```python
# SBSの実装
from sklearn.base import clone
from itertools import combinations
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

class SBS():
    """
    逐次後退選択 (sequential backward selection)
    """
    
    def __init__(self, estimator, k_features, scoring=accuracy_score, test_size=0.25, random_state=1):
        self.scoring = scoring            # 特徴量を評価する指標
        self.estimator = clone(estimator) # 推定器(引数に渡された分類器をdeep copy)
        self.k_features = k_features      # 選択する特徴量の個数
        self.test_size = test_size        # テストデータの割合
        self.random_state = random_state  # 乱数種を固定する値
        
    def fit(self, X, y):
        # トレーニングデータとテストデータに分割
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=self.test_size, random_state=self.random_state)
        
        # すべての特徴量の個数
        dim = X_train.shape[1]
        # 列インデックス
        self.indices_ = tuple(range(dim))
        self.subsets_ = [self.indices_]
        # すべての特徴量でスコアを算出
        score = self._calc_score(X_train, y_train, X_test, y_test, self.indices_)
        # スコアを格納
        self.scores_ = [score]
        # 指定された特徴量の数分繰り返し
        while dim > self.k_features:
            # 空のスコア、列インデックスリスト作成
            scores = []
            subsets = []
            # 列インデックスのうち次元数分選択する全組み合わせで以下の処理を実行
            for p in combinations(self.indices_, r=dim-1):
                # スコア算出、格納
                score = self._calc_score(X_train, y_train, X_test, y_test, p)
                scores.append(score)
                # スコアに対応する列インデックスを格納
                subsets.append(p)
                
            # 特徴量の組み合わせを変えた時に最も高いスコアのインデックスを抽出
            self.indices_ = subsets[np.argmax(scores)]
            self.subsets_.append(self.indices_)
            # 次元を一つ減らす
            dim -= 1
            # 最良のスコアを格納
            self.scores_.append(scores[np.argmax(scores)])
            
        # 最後に格納したスコアが最良
        self.k_score_ = self.scores_[-1]
        
        return self
    
    def transform(self, X):
        # 抽出した特徴量を返す
        return X[:, self.indices_]
        
    def _calc_score(self, X_train, y_train, X_test, y_test, indices):
        # 指定された列番号indicesの特徴量を抽出してモデルに適合
        self.estimator.fit(X_train[:, indices], y_train)
        # テストデータでクラスラベルを予測
        y_pred = self.estimator.predict(X_test[:, indices])
        # 予測値とラベルを比較してスコアを算出
        score = self.scoring(y_test, y_pred)
        return score
```

```python
# knn分類器でSBS実装の効果をみる
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
# インスタンス生成(k=2)
knn = KNeighborsClassifier(n_neighbors=2)
# SBSのインスタンス生成(特徴量の個数が1になるまで選択)
sbs = SBS(knn, k_features=1)
sbs.fit(X_train_std, y_train)
```

```python
# 選択と正解率の関係の可視化
# 特徴量個数リスト（lenで数える）
k_feat = [len(k) for k in sbs.subsets_]
# 横軸を特徴量の個数、縦を正解率(score)でプロット
plt.plot(k_feat, sbs.scores_, marker='o')
plt.ylim([0.7, 1.1])
plt.ylabel('Accuracy')
plt.xlabel('Number of features')
plt.grid()
plt.show()
```

上記のグラフから特徴量を減らすことにより、正解率が向上することがわかる<br>
これは次元の呪いが減少したためと考えられる

```python
# 性能がよかったときの特徴量を確認する
k5 = list(sbs.subsets_[8])
print(df_wine.columns[1:][k5]) # 1つめはclass labelなので2つめから
```

```python
# すべての特徴量を使う場合と、上記の5個を選択した場合で、性能の差を見る
knn.fit(X_train_std, y_train)
# トレーニングデータの正解率
print('Training accuracy:', knn.score(X_train_std, y_train))
# テストデータの正解率
print('Test accuracy:', knn.score(X_test_std, y_test))
```

```python
knn.fit(X_train_std[:, k5], y_train)
# トレーニングデータの正解率
print('Training accuracy:', knn.score(X_train_std[:, k5], y_train))
# テストデータの正解率
print('Test accuracy:', knn.score(X_test_std[:, k5], y_test))
```

学習に利用した特徴量は少ないが、予測精度が2%向上し、トレーニングデータとテストデータとの差も小さくなっており、過学習が軽減されたことがわかる

### ランダムフォレストを用いて特徴量の重要度をみる

ランダムフォレストを用いるとデータが線形分離可能かどうか前提を設けなくて良い<br>
フォレスト内のすべての決定木から計算された不純度の平均的な減少量から特徴量の重要度を測定できる

```python
from sklearn.ensemble import RandomForestClassifier
# Wineデータセットの特徴量の名称
feat_labels = df_wine.columns[1:]
# インスタンス生成(木の数=10000, PCのすべてのコアを計算に使用)
forest = RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1)
# fit
forest.fit(X_train, y_train)
# 重要度を抽出
importances = forest.feature_importances_
# 重要度の降順([::-1])でインデックスを抽出
indices = np.argsort(importances)[::-1]
# 重要度の降順で特徴量の名称と重要度を表示
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
    
plt.title('Feature Importances')
plt.bar(range(X_train.shape[1]), importances[indices], color='lightblue', align='center')
plt.xticks(range(X_train.shape[1]), feat_labels[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.show()
```

```python
# 重要度のしきい値を指定して特徴量を抽出
from sklearn.feature_selection import SelectFromModel
sfm = SelectFromModel(forest, threshold=0.15, prefit=True)
X_selected = sfm.transform(X_train)
X_selected.shape
```

上記の方法を用いると、randomforestを特徴選択器として利用することができる
